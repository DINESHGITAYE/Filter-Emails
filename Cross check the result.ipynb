# Cross check the result:

# Importing clean dataset:

```python
import pandas as pd
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/DINESHGITAYE/Filter-Emails/main/dataset.csv")
```

```python
dataset.head()
```

# Filtering out the Emails that are related to JOB:

```python
dataset[dataset.apply(lambda row: row.astype(str).str.contains('job').any(), axis= 1)]
```

### > Now, see how many Emails are related to JOB:

```python
Emails_related_to_job = dataset[dataset.apply(lambda row: row.astype(str).str.contains('job').any(), axis= 1)]
Emails_related_to_job['Subject'].count()
```

##### There are 177 emails that are related to job

# > To see all these 177 emails which is related to job:

```python
pd.set_option('display.max_rows', Emails_related_to_job.shape[0]+1)
Emails_related_to_job
```
