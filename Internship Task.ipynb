# Internship Task: Filter Emails

# > Extract Emails from Inbox:

```python
import imaplib
import email
import base64
from email.header import decode_header
from getpass import getpass

username = input('Email: ')
password = getpass(prompt='Password: ', stream=None)
```

```python
# create an IMAP4 class with SSL 
imap = imaplib.IMAP4_SSL("imap.gmail.com")
# authenticate
imap.login(username, password)
```

# > Mention how many emails you want to fetch:
##### That is privide value of "N" : 
##### I am extracting first 1000 emails therefore I take N = 1000

```python
status, messages = imap.select("INBOX")
# number of top emails to fetch
N = 1000
# total number of emails
messages = int(messages[0])
```

```python
mail_subject = []
mail_from = []
for i in range(messages, messages-N, -1):
    res, msg = imap.fetch(str(i), "(RFC822)")
    for response in msg:
        if isinstance(response, tuple):
            msg = email.message_from_bytes(response[1])
            subject = decode_header(msg['Subject'])[0][0]
            mail_subject.append(subject)
            from_w = msg.get('From')
            mail_from.append(from_w)
            print("Subject:", subject)
            print("From:", from_w)
imap.close()
imap.logout()
```

# >Data preparation:

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.DataFrame({'From': mail_from, 'Subject': mail_subject})
```

```python
dataset.head()
```

##### > Now remove extra text From "From" column: (remove text between '< >' these signs)

```python
def removal(my_list, position):
    return my_list[position]

dataset["From"] = dataset['From'].str.split('<').apply(removal, position=0)
```

##### > make all the strings in lowercase:
- We are doing this because if we make all the strings in lowercase it is easy to filter out emails using a particular keyword like **'Thank you for applying'**.
- That is emails which have a subject like **"THANK YOU FOR APPLYING"**, "**T**hank **Y**ou **F**or **A**pplying" are treated equally as they converted into **'thank you for applying'**.

```python
dataset["From"] = dataset["From"].str.lower()
dataset["Subject"] = dataset["Subject"].str.lower()
```

```python
dataset[dataset.apply(lambda row: row.astype(str).str.contains('job').any(), axis= 1)]
```
